1) Please plot the noisy data and the polynomial you found (in the same figure). You can use any value of m selected from 2, 3, 4, 5, 6.

2) Plot MSE versus order m for m = 1, 2, 3, 4, 5, 6, 7, 8. Identify the best choice of order m.

The best order m would be 3 since it is the least complex polynomial for that gives the best result.

```
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('seaborn-whitegrid')
import numpy as np

noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.figure(0)
plt.title(f'MSE vs. Order M')

for i in range(1, 9):
  plt.figure(i)
  plt.plot(x,y,'ro', label='Input')

  x = x.ravel()
  y = y.ravel()
  t = np.linspace(-20,5)

  trend = np.polyfit(x,y,i)
  print(f'm={i}: {trend}')
  plt.plot(t, np.polyval(trend, t), label='Trend')
  plt.title(f'M={i}')
  plt.legend()
  mse = ((y - np.polyval(trend,x))**2).mean(axis=0)
  plt.figure(0)
  plt.plot(i, mse, 'ro')
```

3) Change variable noise_scale to 150, 200, 400, 600, 1000 respectively, re-run the algorithm and plot the polynomials with the m found in 2). Discuss the impact of noise scale to the accuracy of the returned parameters. [You need to plot a figure like in 1) for each choice of noise_scale.]

As the noise scale increases, the coefficients of the trendline stray further from the actual coefficients.

```
import matplotlib.pyplot as plt
import matplotlib

plt.style.use('seaborn-whitegrid')
import numpy as np

plt.figure(0)
plt.title(f'MSE vs. noise_scale')

for i in [150, 200, 400, 600, 1000]:
  plt.figure(i)
  number_of_samples = 50
  x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
  y = 5 * x + 20 * x**2 + 1 * x**3 + i*np.random.randn(number_of_samples, 1)
  plt.plot(x,y,'ro', label='Input')

  x = x.ravel()
  y = y.ravel()
  t = np.linspace(-20,5)

  trend = np.polyfit(x,y,3)
  print(f'noise_scale={i}: {trend}')
  plt.plot(t, np.polyval(trend, t), label='Trend')
  plt.title(f'noise_scale={i}')
  plt.legend()
  mse = ((y - np.polyval(trend,x))**2).mean(axis=0)
  plt.figure(0)
  plt.plot(i, mse, 'ro')
```

4) Change variable number_of_samples to 40, 30, 20, 10 respectively, re-ran the algorithm and plot the polynomials with the m found in 2). Discuss the impact of the number of samples to the accuracy of the returned parameters. [You need to plot a figure like in 1) for each choice of number_of_samples.]

The impact of changing the number of samples is relatively limited when compared to introducing more noise. This would mean that it is better to have less data that is better in quality, than more data that is not as good.

```
import matplotlib.pyplot as plt
import matplotlib

plt.style.use('seaborn-whitegrid')
import numpy as np

plt.figure(0)
plt.title(f'MSE vs. number_of_samples')

for i in [40, 30, 20, 10]:
  plt.figure(i)
  noise_scale = 100
  x = 25*(np.random.rand(i, 1) - 0.8)
  y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(i, 1)
  plt.plot(x,y,'ro', label='Input')

  x = x.ravel()
  y = y.ravel()
  t = np.linspace(-20,5)

  trend = np.polyfit(x,y,3)
  print(f'number_of_samples={i}: {trend}')
  plt.plot(t, np.polyval(trend, t), label='Trend')
  plt.title(f'number_of_samples={i}')
  plt.legend()
  mse = ((y - np.polyval(trend,x))**2).mean(axis=0)
  plt.figure(0)
  plt.plot(i, mse, 'ro')
```
